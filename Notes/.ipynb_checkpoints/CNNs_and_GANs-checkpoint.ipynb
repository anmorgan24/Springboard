* The activation map is the value of the filter at every location 
* **When we're dealing with a convolutional layer we want to work with multiple filters because each filter is looking for a specific template or concept or feature in the input volume**
* We can do this for as many filters as we need/want, as shown below:

* If we have six filters (regardless of the filter sizes, which don't have to be the same) then we'll end up with 6 activation maps
* ConvNet is basically a sequence of convolution layers, interspersed with activation functions
* We don't/can't do convolutions without a whole number of slides in each direction

* Output size = $\frac{N-F}{stride+1}$
* For example: 
    * N=7 (image is 7 x 7)
    * F=3 (filter is 3 x 3)
    * `stride1` $\Rightarrow$ (7 - 3)/ 1 + 1 = **5** $\checkmark$
    * `stride2` $\Rightarrow$ (7 - 3)/ 2 + 1 = **3** $\checkmark$
    * `stride3` $\Rightarrow$ (7 - 3)/ 3 + 1 = **2.33** $\boxtimes$

```python
show_image(plt.imread('data/zero_pad.png'), "Zero Padding")
```

```python
(9 - 3)/ 1 + 1
```

**ANSWER:** The output of a 7 x 7 image zero-padded with a 1 pixel border with a 3 x 3 filter applied at stride 1 would be **7 x 7 x the number of filters that you have**.

* Does add some extraneous features to the edges, but we're doing our best to preserve the image
* There are also other methods to try to maximize the output size (minimize information loss), for example: mirroring or extending edge values
* **In general, it is common to see CONV layers with stride 1, filters of size F x F, and zero-padding with:** $\frac{F-1}{2}$
    * For example:
        * F = 3 $\Rightarrow$ zero pad with 1
        * F = 5 $\Rightarrow$ zero pad with 2
        * F = 7 $\Rightarrow$ zero pad with 3

* If we don't use zero padding we're going to really quick shrink the size of the output we have
* This is bad because:
    * Now we're losing out on some of the information
    * Especially losing information around edges (and amplifying information towards center?)

### Pooling layer
* The pooling layer makes the representations smaller and more manageable $\Rightarrow$ fewer parameters
* Operates over each activation map independently

#### Max Pooling

```python
show_image(plt.imread('data/max_pooling.png'), "Max Pooling")
```

* filter size of max pooling = region size
* max pooling takes maximum value of pooled area
* captures most triggering features
* (stride 2 in image above)
* Basically, instead of getting dot product, we take the maximum value
* **In the pooling layers, it is common to set up the strides so that there is no overlap**
    * Part of the point of max pooling is to downsize/consolidate
    * Another option is average pooling, but less common/effective
* Sometimes people will choose to use larger stride values in order to downsize, rather than pooling
* **For pooling layers, common settings:**
    * F = 2, S = 2
    * F - 3, S = 2
    
### Fully Connected Layer (FC Layer)
* Contains neurons that connect to the entire input volume, as with ordinary neural networks
* At the end of any/all Conv/ReLU/Pool layers, you need a FC (Fully-Connected) layer
* Will results in the same 1D output as a vanilla neural network 
* At the last layer we aggregate everything together
* When do you know that you've done enough pooling to do the classification? Trial and error, croos-validation

# GANs
#### Instructor: Ian Goodfellow
### Introduction to Generative Adversarial Networks

#### Adversarial Training
* A phrase whose usage is in flux; a new term that applies to both new and old ideas
* I.G.'s current usage: "Training a model in a worst-case scenario, with inputs chosen by an adversary"
* Examples:
    * An agent playing againt a copy of itsel in a board game (Samuel, 1959)
    * Robust optimization / robust control (e.g. Rustem and Howe 2002)
    * Training neural networks on adversarial examples (Szegedy et al 2013, Goodfellow et al 2014)
    
#### Generative Adversarial Networks
* Both players are neural networks
    * Goal is to try to generate data that resembles the data that was in the training set.
* Worst case input for one network is produced by another network (the adversary)
    * In some cases the worst case input is generated by an optimization algorithm
    * Players are trained to do as well as possible on the worst possible input 
    
* There's two main things that you might ask a generative model to do:
    * 1) Density estimation: we're given a large collection of examples and we want to find the probability density function that describes those examples. 
    * 2) Try to learn a function or a program that can generate more samples from that same training distribution
* The particular approach that generative adversarial networks take to generative modeling is to have two different agents play a game against each other 
    * One of these agents is a **generator network**, which tries to generate data
    * The other agent is a **discriminator network**, that examines data and estimates whether it is real or fake 
    * **The goal of the generator is to fool the discriminator**
    * As both players get better and better at their job over time, **eventually the generator is forced to create data that is as realistic as possible**, data that comes from the same distribution as the training data.
* Police and conterfeiters analogy $\Rightarrow$ The natural equilibrium of the game corresponds to the generator producing perfect samples that come from the same distribution as the training data
    * In other words, the counterfeiter producing counterfeit money that is indistinguishable from real money
    * At that point the discriminator cannot determine between "real" and "fake" and is forced to conclude that any input has 50% probability of being real and 50% probability of being fake
    
#### Minimax Game
* Equilibrium is a saddle point of the discriminator loss
* Resembles Jensen-Shannon divergence
* Generator minimizes the log-probability of the discriminator being correct

**The key insight of Generative Adversarial Networks is to use supervised learning to estimate a ratio that we need to be able to do unsupervised learning.**

#### Non-Convergence
* Optimization algorithms often approach a sadle point or local minimum rather than a global minimum
* Game soliving algorithms may not approach an equilibrium at all
    * Now we're trying to optimize two different loss functions, which might never converge

#### Other pros

* GANs tend to avoid blurriness caused by using MSE
